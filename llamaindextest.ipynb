```python
from llama_index import GPTPandasIndex, LLMPredictor, ServiceContext
from langchain_openai import ChatOpenAI
import pandas as pd

# Load structured data
df = pd.read_csv("students.csv")

# Setup LLM
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
service_context = ServiceContext.from_defaults(llm=llm)

# Create structured data index
pandas_index = GPTPandasIndex(df, service_context=service_context)

# Ask questions
response = pandas_index.query("Who has the highest GPA?")
print(response)
```

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms.openai import OpenAI
from llama_index.experimental.query_engine import PandasQueryEngine
import pandas as pd

# Load your dataframe
df = pd.read_csv("data.csv")

# Define LLM
llm = OpenAI(model="gpt-4-turbo")

# Create a query engine for pandas data
query_engine = PandasQueryEngine(df=df, llm=llm)

# Ask questions about your structured data
response = query_engine.query("What is the average salary of employees?")
print(response)
```

```python
import pandas as pd
from llama_index.llms.openai import OpenAI
from llama_index.query_engine.pandas import PandasQueryEngine

# Load your dataframe
df = pd.read_csv("data.csv")

# Define LLM
llm = OpenAI(model="gpt-4-turbo")

# Create a query engine for structured (tabular) data
query_engine = PandasQueryEngine(df=df, llm=llm)

# Ask questions about the data
response = query_engine.query("Which department has the highest average salary?")
print(response)
```
